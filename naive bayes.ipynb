```python
#count vectorizer

str1= ["hello and welcome to the intel program" , "hi welcome to intel unnati one aprogram", " hi lakshmi pathi sir" , "hai santosh sir"]
```

```python
str1[0].split()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(stop_words='english')
```

```python
op=vect.fit_transform(str1).toarray()
op
```

```python
import pandas as pd
op1= pd.DataFrame(op,columns=vect.get_feature_names_out())
op1
```

```python

url = 'https://raw.githubusercontent.com/arib168/data/main/spam.tsv'

# Read the TSV file directly from the GitHub repository
data1= pd.read_csv(url, sep='\t')  # specify the separator as tab

# Display the first few rows of the dataframe
print(data1.head())
```

```python
data1.to_csv("sample files/data/spam.csv")
```

```python
x=data1.iloc[:,1].values
y=data1.iloc[:,0].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,random_state=0)
```

```python
x_train_vect= vect.fit_transform(x_train)
x_test_vect= vect.transform(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
model1= MultinomialNB()
```

```python
model1.fit(x_train_vect,y_train)
```

```python
y_pred1=model1.predict(x_test_vect)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred1)
```

```python
sample1= ['win free tickets']
sample1 =vect.transform(sample1)
model1.predict(sample1)
```

```python
sample1= ['offer program']
sample1 =vect.transform(sample1)
model1.predict(sample1)

```

```python
from sklearn.pipeline import make_pipeline
model2 = make_pipeline(CountVectorizer(), MultinomialNB())
model2.fit(x_train, y_train) 
```

```python
model2.predict(["hi kill me"])
```

```python
from sklearn.svm import SVC
```

```python
model3 = SVC().fit(x_train_vect,y_train)
```

```python
y_pred3= model3.predict(x_test_vect)
y_pred3
```

```python
op = vect.transform(['gitam university'])
model3.predict(op)
```

```python
accuracy_score(y_test,y_pred3)
```

```python
model4 = make_pipeline(CountVectorizer(),SVC())
```

```python
model4.fit(x_train,y_train)
```

```python
y_pred4= model4.predict(x_test)
y_pred4
```
